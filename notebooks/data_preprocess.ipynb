## Flux en ampère et en méga watt

```python
flows_a_ext = pd.read_csv("_N_flows_a_ext.csv", sep=";")
flows_a_ext.head()
```

```python
flows_a = pd.read_csv("_N_flows_a.csv", sep=";")
flows_a.describe()
```

```python
flows_MW_ext = pd.read_csv("_N_flows_MW_ext.csv", sep=";")
flows_MW_ext.head()
```

```python
flows_MW = pd.read_csv("_N_flows_MW.csv", sep=";")
flows_MW.head()
```

```python
d = {'flows_a_ext':[]}
d['flows_a_ext'] = flows_a_ext.loc[0]
d['flows_a'] = flows_a.loc[0]
d['flows_MW_ext'] = flows_MW_ext.loc[0]
d['flows_MW'] = flows_MW.loc[0]
```

```python
time_step_0 = pd.DataFrame(data=d)
```

```python
time_step_0
```

```python
time_step_0.index
index = np.array(time_step_0.index)
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

p1 = ax.bar(index, time_step_0['flows_MW'])
p2 = ax.bar(index, time_step_0['flows_MW_ext'])

ax.legend((p1,p2), ('flows_MW', 'flows_MW_ext'))
ax.set_xticklabels([]);
```

```python
fig, ax = plt.subplots(figsize=(14, 6))
ax.bar(index, time_step_0['flows_MW'] + time_step_0['flows_MW_ext'])
ax.set_xticklabels([])
plt.plot();
```

Le graphique montre la somme du flux en MW de l'origine et du flux en MW de l'extrémité de chaque ligne pendant le time step 0.

```python
fig, ax = plt.subplots(figsize=(14, 6))

p3 = ax.bar(index, time_step_0['flows_a'])
p4 = ax.bar(index, time_step_0['flows_a_ext'], bottom=time_step_0['flows_a'])

ax.legend((p3, p4), ('flows_a', 'flows_a_ext'))
ax.set_xticklabels([]);
```

```python
fig, ax = plt.subplots(figsize=(19, 6))
sns.heatmap(flows_a, cmap="YlGnBu");

#the evolution of flows in ampere along the time steps
```

Pour les courants, le flux en ampère de l'origine et le flux en ampère de l'extrémité sont de la même valeur et du même signe.

## L'analyse sur le voltage


```python
vext = pd.read_csv("_N_vext.csv", sep=";")
vext.head()
```

```python
vor = pd.read_csv("_N_vor.csv", sep=";")
vor.head()
```

```python
fig, ax = plt.subplots(figsize=(19, 6))
sns.heatmap(vext, cmap="coolwarm");
#the evolution of voltage of extremite of line along the time steps
```

```python
fig, ax = plt.subplots(figsize=(19, 6))
sns.heatmap(vor, cmap="coolwarm");
#the evolution of voltage of origine of line along the time steps
```

```python
fig, ax = plt.subplots(figsize=(19, 6))
sns.heatmap(vext.loc[safe_index], cmap="coolwarm");
#the evolution of voltage of extremite of line along during safe time steps

fig, ax = plt.subplots(figsize=(19, 6))
sns.heatmap(vor.loc[safe_index], cmap="coolwarm");
#the evolution of voltage of origine of line along during safe time steps
```

```python
# calculons la différence de potentiel dans chaque ligne
u = vor - vext
```

```python
fig, ax = plt.subplots(figsize=(19, 6))
sns.heatmap(u, cmap="YlGnBu");
#the evolution of potential difference of line during the whole time steps

fig, ax = plt.subplots(figsize=(19, 6))
sns.heatmap(u.loc[safe_index], cmap="YlGnBu");
#the evolution of potential difference of line during the safe time steps
```

On voit bien qu'il n'y a pas de grande variation au cours de temps pour la différence de potentiel dans chaque ligne. Je pense que ce fait est normal car ce qui est changé au cour du temps est le courant d'électricité qui circule dans les lignes. Avec le résistance constant, la ligne se réchauffe lorsque le courant est grand (selon la d'Ohm et l'effet de Jule).

Cependant, il y a certaines lignes qui possèdent une différence de potentiel assez grande. Je n'ai pas d'idée pourquoi c'est comme ça.
